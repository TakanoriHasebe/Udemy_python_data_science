```python
import numpy as np
```

```python
# リストの作成
my_list1 = [1,2,3,4]
```

```python
# numpy形式に変換
my_array1 = np.array(my_list1)
```

```python
my_array1
```

```python
# 多次元配列を作成する
my_list2 = [11,22,33,44]
my_lists = [my_list1, my_list2]
```

```python
# 多次元配列が作成された
my_lists
```

```python
# numpy型の多次元配列の作成
my_array2 = np.array(my_lists)
```

```python
my_array2
```
