#### <font color='blue'>移動平均について計算をする</font>

移動平均線とはグラフから、10日・20日・50日間という決まった期間を平均してグラフ化する。<br>

```python
ma_day = [10, 20, 50]
for ma in ma_day:
    column_name = 'MA {}'.format(ma)
    # 終値に対して移動平均をとる
    AAPL[column_name] = pd.rolling_mean(AAPL['Close'], ma)
```

```python
AAPL.tail()
```

```python
AAPL[['Close', 'MA 10', 'MA 20', 'MA 50']].plot(subplots=False, figsize=(10, 4))
```

```python
# 株価できになるのは終値を昨日の終値と比較したもの
# pct_changeでそれぞれを比較して、どれくらい変化したかを調べる
AAPL['Daily Return'] = AAPL['Close'].pct_change()
```

```python
AAPL.head()
```

```python
# 前日と比べてどれくらい変化したかが分かるグラフが可視化
AAPL['Daily Return'].plot(figsize=(10, 4), legend=True, linestyle='--', marker='o')
```

```python
# ヒストグラムを作成する
sns.distplot(AAPL['Daily Return'].dropna(), bins = 100, color='purple')
```

```python
# データをとってくる
closing_df = web.DataReader(['AAPL', 'GOOG', 'MSFT', 'AMZN'], 'google', start, end)['Close']
```

```python
closing_df.head()
```

```python
# 終値のパーセンテージで表示
tech_rets = closing_df.pct_change()
```

```python
tech_rets
```

```python
# 相関係数を算出している
sns.jointplot('GOOG', 'GOOG', tech_rets, kind = 'scatter', color = 'seagreen')
```

```python
# microsoftとの相関係数を算出する
sns.jointplot('GOOG', 'MSFT', tech_rets, kind = 'scatter', color = 'seagreen')
```

```python
# 4社分のデータに対して、いっぺんに散布図を書く
sns.pairplot(tech_rets.dropna())
```

```python
# 上側と下側のグラフを変更する。
returns_fig = sns.PairGrid(tech_rets.dropna())
# 各図の指定
returns_fig.map_upper(plt.scatter, color='purple')
returns_fig.map_lower(sns.kdeplot, cmap='cool_d')
returns_fig.map_diag(plt.hist, bins=30)
```

上記から対角線上にヒストグラムが並んでいて<br>
右上に散布図が並んでいる。<br>
左下にカーネル密度関数が表示されている。<br>

```python
# 上側と下側のグラフを変更する。
returns_fig = sns.PairGrid(closing_df)
# 各図の指定
returns_fig.map_upper(plt.scatter, color='purple')
returns_fig.map_lower(sns.kdeplot, cmap='cool_d')
returns_fig.map_diag(plt.hist, bins=30)
```

```python
# 相関係数を数字として求めて、それをそのままプロットする
sns.heatmap(tech_rets.corr(), annot=True)
```

```python
sns.heatmap(closing_df.corr(), annot=True)
```

上記からどこかの株価が上がれば、また別の株価も上がることが分かる。<br>

#### <font color='blue'>以下では、株価のリスク解析について勉強する</font>

```python
# 4社のテクノロジー株の日々の変動からnullを削除
rets = tech_rets.dropna()
```

```python
rets.head()
```

```python

plt.scatter(rets.mean(), rets.std(), alpha=0.5, s = np.pi*20)
```

```python
plt.scatter(rets.mean(), rets.std(), alpha=0.5, s = np.pi*20)
plt.ylim([0.01, 0.025])
plt.xlim([-0.005, 0.01])

plt.xlabel('Expected returns')
plt.ylabel('Risk')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):              
    plt.annotate(label, xy=(x,y), xytext=(0,50),
                 textcoords='offset points', ha = 'right', va = 'bottom',
                 arrowprops = dict(arrowstyle='-', connectionstyle='arc3'))
```
